```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn import svm
from sklearn.model_selection import GridSearchCV
Gclf = GridSearchCV(svm.SVC(gamma='auto'),{'C':[1,10,20], 'kernel':['rbf', 'linear']}, cv=10, return_train_score=False)
Gclf.fit(x_train,y_train)
Gclf.cv_results_
```

```python
dfnew = pd.DataFrame(Gclf.cv_results_)
dfnew
```

```python
Gclf.best_score_
```

```python
Gclf.best_estimator_
```

```python
clf = svm.SVC(C=1, break_ties=False, cache_size=200, class_weight=None, coef0=0.0,
    decision_function_shape='ovr', degree=3, gamma='auto', kernel='linear',
    max_iter=-1, probability=False, random_state=None, shrinking=True,
    tol=0.001, verbose=False)
```

```python
clf.fit(x_train,y_train)
```

```python
support_vector_indices = clf.support_
```

```python
support_vectors_per_class = clf.n_support_
print(support_vectors_per_class)
```

```python
support_vectors = clf.support_vectors_
print(support_vectors)
```

```python
# Visualize support vectors
plt.scatter(x_train[:,0], x_train[:,1])
plt.scatter(support_vectors[:,0], support_vectors[:,1], color='red')
plt.title('Linearly separable data with support vectors')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
y_pred = clf.predict(x_test)
cm = confusion_matrix(y_test,y_pred)
print(cm)
sns.heatmap(cm, annot=True)
```
